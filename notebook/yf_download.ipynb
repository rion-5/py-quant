```python
tickers = ('AAPL', 'AMD', 'AMZN', 'AVGO', 'BA', 'BABA', 'COIN', 'GOOG', 'GOOGL', 'JNJ', 'JPM', 'META', 'MRVL', 'MSFT', 'MSTR', 'ORCL', 'QCOM', 'QQQ', 'TSLA', 'TSM')

download_data = yf.download(tickers, start='2025-01-01',end='2025-01-31', progress=False)
aapl_data = download_data.xs('AAPL', level=1, axis = 1)
print(aapl_data)
```

```python
first_close= aapl_data['Close'].iloc[0]
last_close = aapl_data['Close'].iloc[-1]
price_increase_ratio = (last_close-first_close)/first_close
print(f"(마지막 종가: {last_close } -  첫번째 종가:{ first_close}) / 첫번째 종가:{ first_close} = 상승률 {price_increase_ratio}")
```
